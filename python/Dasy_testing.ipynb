```python
def vector_prep(file, crs, geometry = "", buffer = 0):
    vec_gpd = gpd.read_file(file)

    if geometry != "":
        clip_gpd = gpd.read_file(geometry)

    vec_gpd = vec_gpd.to_crs(crs)
    vec_gpd = vec_gpd.clip(clip_gpd)
    if buffer > 0:
        vec_gpd = vec_gpd.buffer(buffer)

    return vec_gpd

```

```python
import geopandas as gpd
import tobler

# 1. Reproject to Oregon State Plane North (EPSG:2991)
teen_gdf = teen_gdf.to_crs("EPSG:2991")
OSM_clipped = OSM_clipped.to_crs("EPSG:2991")

# 2. Dasymetric interpolation
teen_dasy_interp = tobler.dasymetric.masked_area_interpolate(
    source_df=teen_gdf,
    target_df=OSM_clipped,
    raster=file_path + "isNull_resid1.tif",
    pixel_values=[0],
    extensive_variables=["tot_teen"],
)

# 3. Round and convert to integer
teen_dasy_interp["tot_teen"] = teen_dasy_interp["tot_teen"].round().astype(int)

# 4. Save the result
teen_dasy_interp.to_file(file_path + "teen_dasy_interp3.shp")
print("finished dasy stuff")
```
